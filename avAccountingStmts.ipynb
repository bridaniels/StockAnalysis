# **[Alpha Vantage](https://www.alphavantage.co/documentation/) Accounting Statements**
---

```python
import requests
import json
import pandas as pd
```

```python
API_KEY = 'ALPHA_VANTAGE_API_KEY'
COMPANY = 'GOOG'
```

## Income Statement
---

```python
ACC_DATA = 'INCOME_STATEMENT'
urlIS = 'https://www.alphavantage.co/query?function='+ACC_DATA+'&symbol='+COMPANY+'&apikey='+API_KEY
rIS = requests.get(urlIS).text
dataIS = json.loads(rIS)

print(dataIS)
```

```python
pd.DataFrame(dataIS['annualReports'])
# pd.DataFrame(dataIS['quarterlyReports'])
```

## Balance Sheet
---

```python
ACC_DATA = 'BALANCE_SHEET'
urlBS = 'https://www.alphavantage.co/query?function='+ACC_DATA+'&symbol='+COMPANY+'&apikey='+API_KEY
rBS = requests.get(urlBS)
dataBS = rBS.json()

print(dataBS)
```

```python
# pd.DataFrame(dataBS['annualReports'])
pd.DataFrame(dataBS['quarterlyReports'])
```
