#### Import Libraries 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

# Technical Analysis Library 
import ta
import datetime

from pandas_datareader import data as web
from ta.utils import dropna
```

#### Import Data

### Relative Strength Index (RSI)
---
- not a trend indicator -> momentum indicator 
- measures rise and fall of the stock price 
- compares magnitude of recent gains and losses over time period to measure speed and change of price movements of a security 
- identify overbought/oversold conditions in trading an asset 

```python
rsi = ta.momentum.RSIIndicator(df['Close'], window=14).rsi()

plt.figure(figsize=(16,5))
plt.plot(rsi)

plt.title('{} Relative Strength Index'.format(stock),fontsize=15)
plt.show()
```

### True Strength Indicator (TSI)
---
- shows both trend direction and overbought/oversold conditions 

```python
tsi = ta.momentum.TSIIndicator(df['Close'],window_slow=7, window_fast=21).tsi()

plt.figure(figsize=(16,5))
plt.plot(tsi)

plt.title('{} True Strength Indicator'.format(stock),fontsize=15)
plt.show()
```

### Ultimate Oscillator
---
- Larry Williams (1976): a momentum oscillator designed to capture momentum across three different timeframes

### Kaufman's Adaptive Moving Average (KAMA)
--- 
- accounts for market noise and volatility 
- will closely follow prices when the swings are small and noise is low 
- adjusts when price swings widen and follow prices from a greater distance 
- used to ientify overall trend, time turning points, and filter price movement 

```python
kama = ta.momentum.KAMAIndicator(df['Close'],window=10,pow1=2,pow2=30).kama()

plt.figure(figsize=(16,5))
plt.plot(kama, label='KAMA')
plt.plot(df['Close'],label='Close Prices')

plt.title("{} Kaufman's Adaptive Moving Average".format(stock),fontsize=15)
plt.legend(loc=1)
plt.show()
```

### Rate of Change (ROC)
---
- Momentum: a pure momentum oscillator that measures the percent change in price from one period to the next 
- Compares the current price with the price `n` period ago 
- forms an oscillator that fluctuates above and below the zero line a the Rate-of-Change moves from positive to negative 
- Signals Include: 
    - Center Line Crossovers: prone to whipsaw (especially short-term) -> help identify overall trend 
    - Divergences: fail to foreshadow reversals more of ten than not 
    - Overbought-Oversold Readings: comes naturally to ROC Oscillator 

```python
roc = ta.momentum.ROCIndicator(df['Close'],window=12).roc()

plt.figure(figsize=(16,5))
plt.plot(roc)

plt.title('{} Rate-Of-Change'.format(stock),fontsize=15)
plt.show()
```

### Williams %R
---
- Larry Williams
- inverse of Fast Stochastic Oscillator (aka %R)
- `%R = (HighestHigh - Close)/(HighestHigh - LowestLow) * -100`
- reflects level of the close relative to the highest high for the look-back period 
    - contrast: stochastic oscillator reflects the level of the close relative to the lowest low 
    - %R corrects for the inversion by multiplying the raw value by `-100`
- Fast Stochastic Oscillator and Williams %R produce the exact same lines 
    - scaling is different: Williams %R oscillates from `0 : -100`
    - signals from stochastic oscillator are also applicable to Williams %R
- Readings from `0 : -20` are considered overbought 
- Readings from `-80 : -100` are considered oversold 


```python
william = ta.momentum.WilliamsRIndicator(df['High'],df['Low'],df['Close'],lbp=14).williams_r()

plt.figure(figsize=(16,5))

#plt.plot(stock_sig, label='StochKSignals')
plt.plot(stoch, label='StochK')

plt.plot(william, color='y', label='Williams %R')
plt.axhline(y=0,linewidth=1.5, color='k')
plt.axhline(y=-20, label='Overbought', color='r')
plt.axhline(y=-80, label='Oversold', color='g')

plt.title('{} Williams % R'.format(stock), fontsize=15)
plt.legend(loc=1)
plt.show()
```

### Stochastic RSI
---
- takes advantage of both indicators to create a more sensitive indicator 
- attuded to a specific security's historical performance rather than generalized analysis of price change 

```python
stRSI = ta.momentum.StochRSIIndicator(df['Close'], window=14, smooth1=3, smooth2=3).stochrsi()
stRSI_k = ta.momentum.StochRSIIndicator(df['Close'], window=14, smooth1=3, smooth2=3).stochrsi_k()
stRSI_d = ta.momentum.StochRSIIndicator(df['Close'], window=14, smooth1=3, smooth2=3).stochrsi_d()



plt.figure(figsize=(16,5))
plt.plot(stRSI, label='Stoch RSI')
plt.plot(stRSI_k, label='%K')
plt.plot(stRSI_d, label='%D')

plt.title('{} Stochastic RSI'.format(stock), fontsize=15)
plt.legend(loc=1)
plt.show()
```

### Percentage Point Oscillator (PPO)
---
- momentum oscillator that measures the difference between two moving averages as a percentage of the larger moving average 

```python
ppo_line = ta.momentum.PercentagePriceOscillator(df['Close'],window_slow=26,window_fast=12,window_sign=9).ppo()
ppo_signal = ta.momentum.PercentagePriceOscillator(df['Close'],window_slow=26,window_fast=12,window_sign=9).ppo_signal()
ppo_histogram = ta.momentum.PercentagePriceOscillator(df['Close'],window_slow=26,window_fast=12,window_sign=9).ppo_hist()

plt.figure(figsize=(16,6))
plt.plot(ppo_line, linewidth=2, label='PPO Line')
plt.plot(ppo_signal, label='PPO Signal')
plt.plot(ppo_histogram, label='PPO Histogram')
plt.axhline(y=0, color='k', linestyle='--')

plt.title('{} Percentage Point Oscillator'.format(stock), fontsize=15)
plt.legend(loc=1)
plt.show()
```

### Percentage Volume Oscillator (PVO)
---
- oscillator for volume 
- measures difference between two volume-based moving averages as a percentage of larger moving average 

```python
pvo_line = ta.momentum.PercentageVolumeOscillator(df['Volume'],window_slow=26,window_fast=12,window_sign=9).pvo()
pvo_signal = ta.momentum.PercentageVolumeOscillator(df['Volume'],window_slow=26,window_fast=12,window_sign=9).pvo_signal()
pvo_hist = ta.momentum.PercentageVolumeOscillator(df['Volume'],window_slow=26,window_fast=12,window_sign=9).pvo_hist()



plt.figure(figsize=(16,6))
plt.plot(pvo_line, label='PVO Line', linewidth=2.1)
plt.plot(pvo_signal, label='PVO Signal')
plt.plot(pvo_hist, label='PVO Histogram')
plt.axhline(y=0, color='k', linestyle='--')

plt.legend(loc=1)
plt.title('{} Percentage Volume Oscillator'.format(stock), fontsize=15)
plt.show()
```
